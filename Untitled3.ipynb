```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("customer_dataset_20000.csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
# Gender Count
sns.countplot(data=df, x="Gender")
plt.show()
```

```python
# Age distribution
sns.histplot(df['Age'], bins=30)
plt.show()
```

```python
# Income distribution
sns.histplot(df['AnnualIncome'], bins=30)
plt.show()
```

```python
# Data Cleaning
df.drop_duplicates(inplace=True)

num_cols = ["Age","AnnualIncome","SpendingScore",
            "TransactionsPerMonth","RecencyDays"]

for c in num_cols:
    df[c].fillna(df[c].median(), inplace=True)
```

```python
df["Gender"] = df["Gender"].map({"Male":0, "Female":1})

df = pd.get_dummies(df, columns=["City"], drop_first=True)
```

```python
df.columns
```

```python
features = [
    'Gender','Age','AnnualIncome','SpendingScore',
    'TransactionsPerMonth','RecencyDays'
] + [c for c in df.columns if c.startswith("City_")]

X = df[features]
```

```python
# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

```

```python
X.isnull().sum()
```

```python
print(X_scaled[:5])
print(type(X_scaled))
print(X_scaled.shape)
```

```python
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})
```

```python
df['Gender'] = df['Gender'].fillna(0)
```

```python
X = df[['Gender', 'Age', 'AnnualIncome', 'SpendingScore',
        'TransactionsPerMonth', 'RecencyDays',
        'City_Chennai','City_Delhi','City_Hyderabad','City_Kolkata','City_Mumbai']]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
print(np.isnan(X_scaled).sum())
```

```python
#Elbow method
inertia = []
K = range(2, 11)

for k in K:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X_scaled)
    inertia.append(km.inertia_)

plt.plot(K, inertia, '-o')
plt.title("Elbow Method")
plt.xlabel("K")
plt.ylabel("Inertia")
plt.show()
```

```python
# Silhouette Score
sil_scores = []

for k in K:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X_scaled)
    sil = silhouette_score(X_scaled, labels)
    sil_scores.append(sil)

plt.plot(K, sil_scores, '-o')
plt.title("Silhouette Score")
plt.xlabel("k")
plt.ylabel("Score")
plt.show()
```

```python
# Apply K Means
best_k = 5

kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
df["Cluster"] = kmeans.fit_predict(X_scaled)
```

```python
#PCA Visualization
pca = PCA(n_components=2)
pca_data = pca.fit_transform(X_scaled)

plt.figure(figsize=(8,6))
sns.scatterplot(x=pca_data[:,0], y=pca_data[:,1],
                hue=df["Cluster"], palette="tab10")
plt.title("Customer Segmentation using PCA")
plt.show()
```

```python
#Cluster Profiling
profile = df.groupby("Cluster").agg({
    "Age": "mean",
    "AnnualIncome": "mean",
    "SpendingScore": "mean",
    "TransactionsPerMonth": "mean",
    "RecencyDays": "mean",
    "Gender": "mean"
})

profile
```
